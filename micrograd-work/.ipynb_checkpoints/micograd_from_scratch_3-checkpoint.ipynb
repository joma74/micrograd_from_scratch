```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from graphviz import Digraph

def trace(rootValue):
    nodes, edges = set(), set()
    def build(value):
        if value not in nodes:
            nodes.add(value)
            for term in value._terms:
                edges.add((term, value))
                build(term)
    build(rootValue)
    return nodes, edges

def draw_dot(rootValue, format='svg', rankdir='LR'):
    """
    format: png | svg | ...
    rankdir: TB (top to bottom graph) | LR (left to right)
    """
    assert rankdir in ['LR', 'TB']
    nodes, edges = trace(rootValue)
    dot = Digraph(format=format, graph_attr={'rankdir': rankdir}) #, node_attr={'rankdir': 'TB'})
    
    for n in nodes:
        dot.node(name=str(id(n)), label = "{ %s | data %.4f | grad %.4f }" % (n._label, n.data, n.grad), shape='record')
        if n._op != 'prim':
            dot.node(name=str(id(n)) + n._op, label=n._op)
            dot.edge(str(id(n)) + n._op, str(id(n)))
    
    for n1, n2 in edges:
        dot.edge(str(id(n1)), str(id(n2)) + n2._op)
    
    return dot
```

```python
class Value:
    """ 
        stores a single scalar value and its gradient
    """
    def __init__(self, data: float, _terms=(), _op='prim', _label="?"):
        """ 
        data: a float or integer
        _terms: terms this value combines, starting with self
        _op: operation by how the terms are combined
        _label: label for self
        """
        self.data = data
        self.grad = 0 # by default 0, means initially no effect, does not effect the output. 
                      # So changing the variable does not effect the Loss function
        # internal variables used for micrograd graph construction
        self._terms = set(_terms)
        self._op = _op # the op that produced this node, for graphviz / debugging / etc
        self._label = _label

    def __add__(self, other):
        other = other if isinstance(other, Value) else Value(other)
        out = Value(self.data + other.data, (self, other), '+')
        return out

    def __mul__(self, other):
        other = other if isinstance(other, Value) else Value(other)
        out = Value(self.data * other.data, (self, other), '*')
        return out
        
    def __repr__(self):
        return f"Value(data={self.data}, grad={self.grad}, op={self._op}, label={self._label})"
```

```python
def lol0():
    a = Value(2.0, _label='a')
    b = Value(-3.0, _label='b')
    c = Value(10.0, _label='c')
    e = a * b; e._label='e'
    d = e + c; d._label='d'
    f = Value(-2.0, _label='f')
    L = d * f; L._label='L'
    return L
draw_dot(lol0())
```

```python
# grad(ient): a variable inside the Value class that maintains the derivate of L with respect to that data
# In a neural network settings, L stands for Loss function. Interest lies in the derivative of the Loss function L with 
#   respect to the weight of a neural network.
# ? The leaf nodes will represent the weights of the neural net. But the other leaf nodes will be the data itself. The data 
#   is fixed, so we are not interested in the derivative of that data. But the weights will iterated on using the gradient 
#   information.
# graph with scalar values as inputs that precedes a single output L
# forward pass builds a mathematical expression

# A small amount h (lim->0 t.i. goes to 0) is applied to a - how does it affect L? T.i. what is the derivate of L with 
#   respect to a? T.i. how much has L changed normalized by h?
def lol1():

    h=0.0001
    
    a = Value(2.0, _label='a')
    b = Value(-3.0, _label='b')
    c = Value(10.0, _label='c')
    e = a * b; e._label='e'
    d = e + c; d._label='d'
    f = Value(-2.0, _label='f')
    L = d * f; L._label='L'
    L1 = L.data

    a = Value(2.0 + h, _label='a')
    b = Value(-3.0, _label='b')
    c = Value(10.0, _label='c')
    e = a * b; e._label='e'
    d = e + c; d._label='d'
    f = Value(-2.0, _label='f')
    L = d * f; L._label='L'
    L2 = L.data

    print('The derivative of L with respect to a is', (L2 - L1)/h)

lol1()
```

```python
# Start manually filling the grads backward
#
# 1. What is the derivative of L with respect to L
# L = L
# dL/dL = ?
# formula: (f(x+h) - f(x)) / h
# ((L + h) - L) / h = h / h = 1
# So dL/dL is 1, hence grad of L is 1
#
# Inline gradient step to numerically verify dL/dL=1
def lol2():

    h=0.0001
    
    a = Value(2.0, _label='a')
    b = Value(-3.0, _label='b')
    c = Value(10.0, _label='c')
    e = a * b; e._label='e'
    d = e + c; d._label='d'
    f = Value(-2.0, _label='f')
    L = d * f; L._label='L'
    L1 = L.data

    a = Value(2.0, _label='a')
    b = Value(-3.0, _label='b')
    c = Value(10.0, _label='c')
    e = a * b; e._label='e'
    d = e + c; d._label='d'
    f = Value(-2.0, _label='f')
    L = d * f; L._label='L'
    L += h
    L2 = L.data

    print('The derivative of L with respect to L is', (L2 - L1)/h)

lol2()
```

```python
# Provide dL/dL=1 to the variable and draw
def lol3():
    a = Value(2.0, _label='a')
    b = Value(-3.0, _label='b')
    c = Value(10.0, _label='c')
    e = a * b; e._label='e'
    d = e + c; d._label='d'
    f = Value(-2.0, _label='f')
    L = d * f; L._label='L'
    # manually calculated grads
    L.grad = 1.0
    return L
draw_dot(lol3())
```

```python
# 2. What is the derivative of L with respect to d?
# L = d * f
# dL/dd = ?
# formula: (f(x+h) - f(x)) / h
# ((d + h) * f) - (d * f) / h
# ((d * f) + (h * f) - (d * f)) / h
# h * f / h
# f
# So dL/dd is f, hence grad of L is -2.0
# 3. What is the derivative of L with respect to f?
# L = d * f
# dL/df = ?
# formula: (f(x+h) - f(x)) / h
# ((d * (f + h)) - (d * f) / h
# ((d * f + d * h) - (d * f)) / h
# d * h / h
# d
# So dL/df is d, hence grad of L is 4.0
def lol4():
    a = Value(2.0, _label='a')
    b = Value(-3.0, _label='b')
    c = Value(10.0, _label='c')
    e = a * b; e._label='e'
    d = e + c; d._label='d'
    f = Value(-2.0, _label='f')
    L = d * f; L._label='L'
    # manually calculated grads
    L.grad = 1.0
    d.grad = f.data
    f.grad = d.data
    return L
draw_dot(lol4())
```

```python
# 3. What is the derivative of L with respect to c? 
# KNOW
#   dL/dd
# WANT
#   dL/dc
#   How does c effect L trough d? If we know how d effects L and how c effects d, then we know how c effects L.
#   How c effects d is the local derivative
# WANT
#   dd/dc
#   d = c + e
#   formula: (f(x+h) - f(x)) / h
#   ((c + h) + e) - (c + e)) / h
#   (h + e) - e / h
#   1
# So dd/dc is 1, hence grad of c with respect to d is 1.0
# WANT
#   dL/dc = (dL/dd) * (dd/dc)
#     dL/dd = -2.0 = d.grad
#     dd/dc = 1
#   dL/dc = d.grad * 1
# 4. What is the derivative of L with respect to e? 
# WANT
#   dL/de = (dL/dd) * (dd/de)
#     dL/dd = -2.0 = d.grad
#     dd/de = 1
#   dL/de = d.grad * 1
def lol5():
    a = Value(2.0, _label='a')
    b = Value(-3.0, _label='b')
    c = Value(10.0, _label='c')
    e = a * b; e._label='e'
    d = e + c; d._label='d'
    f = Value(-2.0, _label='f')
    L = d * f; L._label='L'
    # manually calculated grads
    L.grad = 1.0
    d.grad = f.data
    f.grad = d.data
    c.grad = (d.grad * 1)
    e.grad = (d.grad * 1)
    return L
draw_dot(lol5())
```

```python
# 5. What is the derivative of L with respect to a? 
# KNOW
#   dL/de
# WANT
#   dL/da
#   How does a effect L trough e? If we know how e effects L and how a effects e, then we know how a effects L.
#   How a effects e is the local derivative
# WANT
#   de/da
#   e = a * b
#   formula: (f(x+h) - f(x)) / h
#   ((a + h) * b) - (a * b)) / h
#   (a * b + h * b) - (a * b) / h
#   b
# So de/da is b, hence grad of a with respect to e is b
# WANT
#   dL/da = (dL/de) * (de/da)
#     dL/de = -2.0 = e.grad
#     de/da = b
#   dL/da = e.grad * b
# 6. What is the derivative of L with respect to b? 
# WANT
#   dL/db = (dL/de) * (de/db)
#     dL/de = -2.0 = e.grad
#     de/db = a
#   dL/da = e.grad * a
def lol6():
    a = Value(2.0, _label='a')
    b = Value(-3.0, _label='b')
    c = Value(10.0, _label='c')
    e = a * b; e._label='e'
    d = e + c; d._label='d'
    f = Value(-2.0, _label='f')
    L = d * f; L._label='L'
    # manually calculated grads
    L.grad = 1.0
    d.grad = f.data
    f.grad = d.data
    c.grad = (d.grad * 1)
    e.grad = (d.grad * 1)
    a.grad = (e.grad * b.data)
    b.grad = (e.grad * a.data)
    return L
draw_dot(lol6())
```

```python
# Now we nudge our inputs(t.i. leaf nodes we have control over) to try to get L up.
# Hence we must go in the direction of our gradient.
# We nudge the leaf nodes in the step size 0.01 and then rerun the forward pass.
# This process is one step of an optimiziation.
def lol7():
    a = Value(2.0, _label='a')
    b = Value(-3.0, _label='b')
    c = Value(10.0, _label='c')
    e = a * b; e._label='e'
    d = e + c; d._label='d'
    f = Value(-2.0, _label='f')
    L = d * f; L._label='L'
    # manually calculated grads
    L.grad = 1.0
    d.grad = f.data
    f.grad = d.data
    c.grad = (d.grad * 1)
    e.grad = (d.grad * 1)
    a.grad = (e.grad * b.data)
    b.grad = (e.grad * a.data)
    # We nudge the leaf nodes in the step size ...
    stepsize = 0.01
    a.data += stepsize * a.grad
    b.data += stepsize * b.grad
    c.data += stepsize * c.grad
    f.data += stepsize * f.grad
    # ... and then rerun the forward pass
    e = a * b; e._label='e'
    e.grad = (d.grad * 1)
    d = e + c; d._label='d'
    d.grad = f.data
    L = d * f; L._label='L'
    L.grad = 1.0
    return L
draw_dot(lol7())
```
