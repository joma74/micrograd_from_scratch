```r
# Replay of https://www.youtube.com/watch?v=C4N3_XJJ-jU&t=613s
# Import and clean up the data
# Data from https://archive.ics.uci.edu/static/public/45/heart+disease.zip
```

```r
data <- read.csv(file = "./processed.cleveland.data", header = FALSE)
```

```r
head(data)
```

```r
colnames(data) <- c("age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang",
                    "oldpeak", "slope", "ca", "thal", "hd")
```

```r
head(data)
```

```r
str(data) # describes the structure of the data
```

```r
data[data == "?"] <- NA # Change the "?" to NAs
```

```r
data[data == "?"]
```

```r
data[data == "NA"]
```

```r
data[data$sex == 0,]$sex <- "F" # convert the 0s in sex to F
data[data$sex == 1,]$sex <- "M" # convert the 1s in sex to M
data$sex <- as.factor(data$sex)
```
