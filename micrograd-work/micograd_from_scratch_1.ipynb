```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return 3*x**2 - 4*x + 5
```

```python
f(3.0)
```

```python
xs = np.arange(-5, 5, 0.25)
ys=f(xs)
plt.plot(xs, ys);  plt.grid();
```

```python
h = 0.0000001 # a tiny diff
x = 2/3 # the inflection point, hence the derivative y at this point x is ~ 0
(f(x+h) - f(x)) / h # differentiate the rise t.i. y for a tiny change h in the run t.i. x
```

```python
h = 0.0001 # a tiny diff
# inputs
a = 2.0
b = -3.0
c = 10.0
d1 = a*b + c
# the derivate of d with respect differntiate by a
a += h
d2 = a*b + c
print('d1', d1)
print('d2', d2)
print ('slope or grad(ient)', (d2 - d1)/h)
```
